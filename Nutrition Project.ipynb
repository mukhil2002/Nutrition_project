```python
def categorize_obesity(mean_estimate):
    if mean_estimate >= 30:
        return 'High'
    elif mean_estimate >= 25 and mean_estimate < 30:
        return 'Moderate'
    else:
        return 'Low'

df_obesity['obesity_level'] = df_obesity['Mean_Estimate'].apply(categorize_obesity)
```

```python
# TABLE CREATION
```

```python
cursor.execute("""
CREATE TABLE obesity (
    id INT AUTO_INCREMENT PRIMARY KEY,
    Region VARCHAR(255),
    Gender VARCHAR(50),
    Year INT,
    LowerBound FLOAT,
    UpperBound FLOAT,
    Mean_Estimate FLOAT,
    Country VARCHAR(255),
    age_group VARCHAR(50),
    CI_Width FLOAT,
    obesity_level VARCHAR(50)
);
""")
```

```python
cursor.execute("""
CREATE TABLE malnutrition (
    id INT AUTO_INCREMENT PRIMARY KEY,
    Region VARCHAR(255),
    Gender VARCHAR(50),
    Year INT,
    LowerBound FLOAT,
    UpperBound FLOAT,
    Mean_Estimate FLOAT,
    Country VARCHAR(255),
    age_group VARCHAR(50),
    CI_Width FLOAT,
    malnutrition_level VARCHAR(50)
);
""")
```

```python
obesity_records = df_obesity.to_dict('records')
```

```python
malnutrition_records = df_malnutrition.to_dict('records')
```

```python
myconnection.commit()
```

```python
insert_obesity_query = """
INSERT INTO obesity (
    Region, Gender, Year, LowerBound, UpperBound,
    Mean_Estimate, Country, age_group, CI_Width, obesity_level
) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""
for record in obesity_records:
    obesity_values = (
        record['Region'],
        record['Gender'],
        record['Year'],
        record['LowerBound'],
        record['UpperBound'],
        record['Mean_Estimate'],
        record['Country'],
        record['age_group'],
        record['CI_Width'],
        record['obesity_level']
    )
    cleaned_obesity_values = [None if pd.isna(item) else item for item in obesity_values]
    cursor.execute(insert_obesity_query, tuple(cleaned_obesity_values))
myconnection.commit()
print(f"Successfully inserted {len(obesity_records)} rows into 'obesity'.")
```

```python
insert_malnutrition_query = """
INSERT INTO malnutrition (
    Region, Gender, Year, LowerBound, UpperBound,
    Mean_Estimate, Country, age_group, CI_Width, malnutrition_level
) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""
for record in malnutrition_records:
    malnutrition_values = (
        record['Region'],
        record['Gender'],
        record['Year'],
        record['LowerBound'],
        record['UpperBound'],
        record['Mean_Estimate'],
        record['Country'],
        record['age_group'],
        record['CI_Width'],
        record['malnutrition_level']
    )
    cleaned_malnutrition_values = [None if pd.isna(item) else item for item in malnutrition_values]
    cursor.execute(insert_malnutrition_query, tuple(cleaned_malnutrition_values))
myconnection.commit()
print(f"Successfully inserted {len(malnutrition_records)} rows into 'malnutrition'.")
```

```python
## COMBINED QUERIES
```

```python
# 1. Obesity vs malnutrition comparison by country (any 5 countries - e.g., India, USA, Brazil, Nigeria, Germany)
pd.read_sql_query("""
SELECT
    obesity.Country,
    obesity.Year,
    obesity.Mean_Estimate AS Obesity_Mean_Estimate,
    malnutrition.Mean_Estimate AS Malnutrition_Mean_Estimate
FROM
    obesity
JOIN
    malnutrition ON obesity.Country = malnutrition.Country AND obesity.Year = malnutrition.Year
WHERE
    obesity.Country IN ('India', 'United States of America', 'Brazil', 'Nigeria', 'Germany')
ORDER BY
    obesity.Country, obesity.Year;
""", myconnection)
```

```python
# 2. Gender-based disparity in both obesity and malnutrition (average across all years)
# Obesity:
pd.read_sql_query("""
SELECT
    Gender,
    AVG(Mean_Estimate) AS Average_Obesity_Estimate
FROM
    obesity
GROUP BY
    Gender
ORDER BY
    Gender;
""", myconnection)
```

```python
# 2.
# Malnutrition

pd.read_sql_query("""
SELECT
    Gender,
    AVG(Mean_Estimate) AS Average_Malnutrition_Estimate
FROM
    malnutrition
GROUP BY
    Gender
ORDER BY
    Gender;
""", myconnection)
```

```python
# 3. Region-wise average estimates side-by-side (Africa and America)
pd.read_sql_query("""
SELECT
    obesity.Region,
    AVG(obesity.Mean_Estimate) AS Average_Obesity_Estimate,
    AVG(malnutrition.Mean_Estimate) AS Average_Malnutrition_Estimate
FROM
    obesity
JOIN
    malnutrition ON obesity.Region = malnutrition.Region AND obesity.Year = malnutrition.Year
WHERE
    obesity.Region IN ('Africa', 'Americas')
GROUP BY
    obesity.Region
ORDER BY
    obesity.Region;
""", myconnection)
```

```python
# 4. Countries with obesity up & malnutrition down

pd.read_sql_query("""
SELECT
    obesity.Country,
    obesity.Year,
    obesity.obesity_level,
    malnutrition.malnutrition_level,
    obesity.Mean_Estimate AS Obesity_Mean_Estimate,
    malnutrition.Mean_Estimate AS Malnutrition_Mean_Estimate
FROM
    obesity
JOIN
    malnutrition ON obesity.Country = malnutrition.Country
WHERE
    obesity.obesity_level = 'High' AND malnutrition.malnutrition_level = 'Low'
ORDER BY
    obesity.Country;
""", myconnection)
```

```python
# 5. Age-wise trend analysis for both obesity and malnutrition (average over years per age group)
# Obesity
pd.read_sql_query("""
SELECT
    Year,
    age_group,
    AVG(Mean_Estimate) AS Average_Obesity_Estimate
FROM
    obesity
GROUP BY
    Year, age_group
ORDER BY
    Year ASC, age_group;
""", myconnection)
```

```python
# 5. 
# Malnutrition
pd.read_sql_query("""
SELECT
    Year,
    age_group,
    AVG(Mean_Estimate) AS Average_Malnutrition_Estimate
FROM
    malnutrition
GROUP BY
    Year, age_group
ORDER BY
    Year ASC, age_group;
""", myconnection)
```
